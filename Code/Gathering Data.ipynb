```python
import requests
response = requests.get("https://ballotpedia.org/List_of_current_members_of_the_U.S._Congress")
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")
senators = soup.find_all("tbody")[3]
representatives = soup.find_all("tbody")[6]
senators_links = []
representatives_links = []
```

Gathering all Senators Ballotopedia Page

```python
for row in senators.find_all("tr"):
  link = row.find_all("a")[1]["href"]
  senators_links.append(link)
  print(link)
```

Parsing through each page gathering data and their twitter, giving preference to campaign twitter.

```python
senator_df = pd.DataFrame(columns = ["Name", "Party", "State", "Start", "Twitter"])
link = senators_links[0]
i=0
for link in senators_links:
  response = requests.get(link)
  soup = BeautifulSoup(response.text, "html.parser")
  infobox = soup.find("div", class_ = "infobox person")
  name = infobox.find("div").text
  party = infobox.find("div", class_ = "widget-row value-only white").text.strip()
  twitter_link = ""
  for p in soup.find("div", class_ = "mw-parser-output").find_all("p"):
    if "assumed" in p.text and name in p.text:
      sentence = p.text
      start = sentence.find("assumed office on") + 18
      start_date = sentence[start:sentence.find('.', start)]
      start = sentence.find("from") + 5
      state = sentence[start: sentence.find(".", start)]
    if "Campaign Twitter" in p.text and twitter_link == "":
      twitter_link = p.find("a")["href"]
    if "Official Twitter" in p.text and twitter_link == "":
      twitter_link = p.find("a")["href"]
    if "Personal Twitter" in p.text and twitter_link == "":
      twitter_link = p.find("a")["href"]
  body = "Senate"
  twitter_user = twitter_link[twitter_link.find("/", 10) + 1:]
  new_addition = pd.DataFrame({"Name":[name], "Party": [party], "State": [state], "Start" : [start_date], "Twitter": [twitter_user] })
  senator_df = pd.concat([senator_df, new_addition], ignore_index = True)
  i+=1
  if i%10 == 0:
    print(str(i) + "/100")
senator_df
```

```python
senator_df.to_csv("/Senators_Twitter.csv")
```

Gathering Representatives 

```python
for row in representatives.find_all("tr"):
  link = row.find_all("a")[1]["href"]
  representatives_links.append(link)
  print(link)
```

Gathering same data as Senators with district too

```python
representative_df = pd.DataFrame(columns = ["Name", "Party", "State", "District", "Start", "Twitter"])
link = representatives_links[0]
i=0
for link in representatives_links:
  response = requests.get(link)
  soup = BeautifulSoup(response.text, "html.parser")
  infobox = soup.find("div", class_ = "infobox person")
  name = infobox.find("div").text
  party = infobox.find("div", class_ = "widget-row value-only white").text.strip()
  twitter_link = ""
  for p in soup.find("div", class_ = "mw-parser-output").find_all("p"):
    if "assumed" in p.text and name in p.text:
      sentence = p.text
      start = sentence.find("assumed office on") + 18
      start_date = sentence[start:sentence.find('.', start)]
    if "member" in p.text:
      start = sentence.find("representing") + 13
      end =  sentence.find("'", start)
      state = sentence[start:end]
      district = sentence[sentence.find(" ", end) + 1]
      if sentence[sentence.find(" ", end) + 2].isnumeric():
        district += sentence[sentence.find(" ", end) + 2]
    if "Campaign Twitter" in p.text and twitter_link == "":
      twitter_link = p.find("a")["href"]
    if "Official Twitter" in p.text and twitter_link == "":
      twitter_link = p.find("a")["href"]
    if "Personal Twitter" in p.text and twitter_link == "":
      twitter_link = p.find("a")["href"]
  body = "House of Representatives"
  twitter_user = twitter_link[twitter_link.find("/", 10) + 1:]
  new_addition = pd.DataFrame({"Name":[name], "Party": [party], "State": [state], "District":[district], "Start" : [start_date], "Twitter": [twitter_user] })
  representative_df = pd.concat([representative_df, new_addition], ignore_index = True)
  i+=1
  if i%10 == 0:
    print(str(i) + "/441")
representative_df
```

```python
representative_df.to_csv("/representative2.csv")
```

Cleaning data for Senators with spot fix and DateTime

```python
senate_df = pd.read_csv("https://raw.githubusercontent.com/kyleschmoyer/DataSci/main/Senators_Twitter.csv")
senate_df = senate_df.drop("Unnamed: 0", axis = 1)
senate_df.loc[53, "Twitter"] = "SenatorRicketts"
senate_df.loc[90, "Twitter"] = "timkaine"
senate_df["Start"] = pd.to_datetime(senate_df["Start"], format="%B %d, %Y")
senate_df["Start"] = senate_df["Start"].dt.date
senate_df
```

Gathering 1000 tweets from each senator along with data up to when they took office

```python
import snscrape.modules.twitter as sntwitter
senate_tweets = pd.DataFrame(columns = ["User", "Text", "Hashtag", "Lang", "Likes", "Retweets", "Quotes", "MentionedUsers", "Date"])


for j in range(len(senate_df)):
  user = senate_df.loc[j, "Twitter"]
  since_date = senate_df.loc[j, "Start"]
  search_query = f"from:{user} since:{since_date}"

  for i, tweet in enumerate(sntwitter.TwitterSearchScraper(search_query).get_items()):
    if i == 1000:
      break
    new_addition = pd.DataFrame({"User":[user], "Text": [tweet.rawContent], "Hashtag": [tweet.hashtags], "Lang" : [tweet.lang], "Likes": [tweet.likeCount],"Retweets": [tweet.retweetCount], "Quotes":[tweet.quoteCount], "MentionedUsers":[tweet.mentionedUsers], "Date": [tweet.date]})
    senate_tweets = pd.concat([senate_tweets, new_addition], ignore_index = True)
  if j%10 == 0:
    print(str(j) + "/100")

senate_tweets
```

```python
senate_tweets.to_csv("/senate_tweets.csv")
```
